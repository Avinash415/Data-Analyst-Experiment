```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data using NumPy
x = np.linspace(0, 10, 100)  # Create an array of 100 points from 0 to 10
y = np.sin(x)  # Compute the sine values for each x

# Create an interactive plot
plt.figure(figsize=(8, 6))  # Set the figure size

# Plot the data with specific properties
plt.plot(x, y, label='Sin(x)', color='blue', linestyle='-', linewidth=2, marker='o', markersize=8)

# Set properties of the plot
plt.title('Interactive Chart with Matplotlib')  # Set the title
plt.xlabel('X-axis')  # Set the label for the x-axis
plt.ylabel('Y-axis')  # Set the label for the y-axis
plt.grid(True)  # Add grid lines
plt.legend()  # Display the legend
plt.tight_layout()  # Adjust layout for better appearance

# Show the interactive plot
plt.show()
```
